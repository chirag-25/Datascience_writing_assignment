```python
x = data['Year']
y = data['GDP at Factor Cost']

fig, ax = plt.subplots(figsize=(15, 10))

# Plot the data as a line chart
ax.plot(x, y)

# Add chart title and axis labels
ax.set_title('GDP at Factor Cost Over Time')
ax.set_xlabel('Year')
ax.set_ylabel('GDP at Factor Cost')
plt.xticks(rotation=90)
# Show the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Select the columns to plot
x = data['Year']
y_cols = data.columns[1:6] # Select the next 5 columns after 'Year'

# Define a list of colors for each line
colors = ['blue', 'red', 'green', 'orange', 'purple']

# Create a plot with multiple lines
plt.figure(figsize=(15,10)) # Set the figure size
for i, col in enumerate(y_cols):
    plt.plot(x, data[col], label=col, color=colors[i])

# Add chart title and axis labels
plt.title('Macro Economic Indicators Over Time')
plt.xlabel('Year')
plt.ylabel('Value')

# Add a legend
plt.legend()
plt.xticks(rotation = 90)
# Show the chart
plt.show()
```
